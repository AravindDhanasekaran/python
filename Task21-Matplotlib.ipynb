```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(0,19,2)
print(x)
print(y)
```

```python
#scatter graph
plt.scatter(x,y,c='b',marker='+')
```

```python
#line graph
plt.plot(x,y,"r")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Test graph")
```

```python
#labeling
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
#doubt text(0, 0.5, 'Y-axis')
```

```python
x=np.arange(0,30,0.1)
y=np.sin(x)
```

```python
#sine graph 
plt.plot(x,y,"r")
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,"r")
```

```python
#A survey of 145 people asked them "Which is the nicest fruit?":
#bar graph
x=["Apple","Orange","Banana","Kiwifruit","blueberry","Grapes"]
y=[35,30,10,25,40,5]
```

```python
plt.bar(x,y)
```

```python
#pie chart
x = [1000, 200, 300, 50]
labels = ['China', 'US', 'Russia', 'India']
colors = ['r', 'g', 'b', 'y']

explode = (0.2, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(x, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%',shadow=True)
plt.title('Corona Virus Affected people')
```
