```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.tail(10)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#jointplot 
#it helps us to give visualization for univariate or bivariate functions
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
#pair plot 
#when we have multiple feature to calculate such that 2 indepent feature
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="smoker")
```

```python
#get count
df["smoker"].value_counts
```

```python
#countplot
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
#box plot used to represent the percentiles
sns.boxplot('total_bill',data=df)
```

```python
#violin plot
sns.violinplot(x='total_bill',y="day",data=df)
```
